<h1><strong>Step 3: Visualization</strong></h1>

<h1><strong>Step 4: Model building</strong></h1>

```python
import torch
import torch.nn as nn

X = torch.tensor(df[["hours_studied", "sleep_hours"]].values, dtype=torch.float32)
y = torch.tensor(df["score"].values, dtype=torch.float32).view(-1, 1)

model = nn.Linear(2, 1)

criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# train loop
for epoch in range(200):
    pred = model(X)  
    # loss cal
    loss = criterion(pred, y)
    # clr accumulated data
    optimizer.zero_grad()
    # cal gradient , backward pass
    loss.backward()
    # updating weigths & bia (ie model parameter)
    optimizer.step()

print("Training complete")
```
